```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
# lr.predict(x_test.iloc[0].values.reshape(1,1))
lr.predict(x_test.iloc[[1]])  # Using double brackets to keep as a DataFrame
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
#slope
m=lr.coef_
```

```python
# y-intercept
c=lr.intercept_
```

```python
# y=mx+c

m*8.58+c  # package for a student with 8.58 cgpa(this is how is algo works bts using this equation of st line to find a line that best fits the given data and then make predictions out of it)
```
